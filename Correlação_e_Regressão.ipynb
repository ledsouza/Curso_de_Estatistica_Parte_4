### Prepare os dados para estimar um modelo de regressão linear simples

```python
Y = dataset.Y
X = sm.add_constant(dataset.X)
```

### Estime o modelo de regressão linear simples

```python
regressao_linear = sm.OLS(Y, X, missing='drop').fit()
```

### Visualize o resultado da regressão

```python
print(regressao_linear.summary())
```

### Obtenha o $Y$ previsto

```python
dataset['Y_previsto'] = regressao_linear.predict()
dataset.head()
```

### Utilizando nosso modelo estimado para fazer previsões.
### Qual seria o consumo de cerveja para um dia com temperatura média de 42° C?

```python
print(f'O consumo de cerveja para um dia com temperatura média de 42° C é {regressao_linear.predict([1, 42])[0]:.2f} litros')
```

### Obtenha os resíduos da regressão

```python
dataset['Residuos'] = regressao_linear.resid
dataset.head()
```

### Plote um gráfico de dispersão dos resíduos da regressão contra o $Y$ previsto

```python
ax = sns.scatterplot(x=dataset['Y_previsto'], y=dataset.Residuos)
ax.figure.set_size_inches(12, 6)
ax.set_title('Resíduos vs Y previsto', fontsize=18)
ax.set_xlabel('Y previsto', fontsize=14)
ax.set_ylabel('Resíduos', fontsize=14)
ax
```

### Obtenha o QQPlot dos resíduos

```python
from scipy.stats import probplot
import matplotlib.pyplot as plt

(_, (_, _, _)) = probplot(dataset.Residuos, plot = plt)
```

### <font color='red'>A partir dos outputs acima, qual conclusão podemos tirar sobre o comportamento dos resíduos?</font>

O plot de "Residuos vs Y previsto" sugere que não haja heterocedasticidade nos dados. Ou seja, a dispersão dos dados não é uniforme ao longo da linha de regressão.
O "Probability Plot", por sua vez, sugere que não haja normalidade na distribuição dos resíduos.

### Obtenha o R² da regressão pelo método da soma dos quadrados

# $$R^2 = \frac{SQR}{SQT}$$

```python
SQR = regressao_linear.ess
SQE = regressao_linear.ssr
SQT = SQR + SQE
R2 = SQR / SQT
R2
```
